```python
import pandas as pd
import numpy as np

pd.__version__
```

```python
name_file = "Dane_ofert_Otodom_2024_08_28.csv"
```

```python
df = pd.read_csv(name_file, index_col=0)
```

```python
df.info()
```

```python
df.columns = [col.lower() for col in df.columns]
df = df.rename(columns={'price [pln]': 'price', 'surface [m2]': 'surface ', 'price_per_m2 [pln/m2]': 'price_per_m2'})
df.head()
```

```python
df[:].replace('Unknow', np.nan, inplace=True, regex=False)
df[:].replace('Eror value', np.nan, inplace=True)
df['district'] = df['district'].str.replace(' ', "_", regex=False)
# df['offer_type'] = df['offer_type'].str.replace(' ', "_")
df['street'] = df['street'].str.replace('ul. ', "", regex=False)
df['street'] = df['street'].str.replace('al. ', "", regex=False)
df['street'] = df['street'].str.replace(' ', "_", regex=False)
df.to_csv(name_file)
df.head(20)
```
