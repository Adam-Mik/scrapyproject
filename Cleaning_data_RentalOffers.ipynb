```python
import pandas as pd
import numpy as np

pd.__version__
```

```python
name_file = "Dane_ofert_wynajem_Rzeszow_Otodom_2024_08_31.csv"
```

```python
df = pd.read_csv(name_file, index_col=0)
df
```

```python
df.shape
```

```python
df.columns = [col.lower() for col in df.columns]
df = df.rename(columns={'price [pln]': 'price', 'surface [m2]': 'surface ', 'rent [pln/month]': 'rent'})
df.head()
```

```python
df['rent'].replace('Unknow', np.nan, inplace=True)
df['rent'].replace('', np.nan, inplace=True)
df['district'] = df['district'].str.replace(' ', "_", regex=False)
# df['offer_type'] = df['offer_type'].str.replace(' ', "_", regex=False)
df['street'] = df['street'].str.replace('ul. ', "", regex=False)
df['street'] = df['street'].str.replace('al. ', "", regex=False)
df['street'] = df['street'].str.replace(' ', "_", regex=False)
```
