**Save to html best offers**

```python
with open('Interesting_offers.html', 'w', encoding='utf-8') as f:
    f.write('<h1>Spis treści</h1>\n')
    f.write('<ul>\n')  

    # Tworzymy linki do poszczególnych dzielnic
    for distr in district_name:
        # linki do poszczególnych dzielnic
        f.write(f'<li><a href="#{distr}">{distr}</a></li>\n')  # Link do dzielnicy
    
    f.write('</ul>\n')  # Zamykamy listę
    
    # Teraz dodajemy wyniki dla każdej dzielnicy
    for distr in district_name:
        x = df[(df['district'] == distr) & 
               (
                (df['price'] < 0.88 * df[df['district'] == distr]['price'].mean()) |
               (df['price_per_m2'] < 0.88 * df[df['district'] == distr]['price_per_m2'].mean())
               )
              ][['district', 'price', 'price_per_m2', 'number_of_room', 'surface ', 'link']]
        
        # Dodajemy sekcję dla danej dzielnicy z odpowiednim identyfikatorem
        f.write(f'<h2 id="{distr}">----------{distr}----------</h2>\n')
        f.write(x.to_html(index=False))
        f.write('<br><br>')

print("Wyniki zapisane do 'wyniki.html' z dodanym spisem treści.")
```
